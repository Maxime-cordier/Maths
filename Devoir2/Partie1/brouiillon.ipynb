1. Ecrire un programme générant N nombres aléatoires pour une loi discrète et une loi continue. Représenter graphiquement des séquences de nombres en faisant varier N.

Loi discrète : Loi Binomial

```python
def random_N_Binomiale(N):
    return [random.randint(0, 1) for _ in range(N)]
```

```python
N10 = random_N_Binomiale(10)    
N100 = random_N_Binomiale(100)
N1000 = random_N_Binomiale(1000)
N5000 = random_N_Binomiale(5000)
N10000 = random_N_Binomiale(10000)
# create data
df = pd.DataFrame([["10", N10.count(0), N10.count(1)], 
                   ["100", N100.count(0), N100.count(1)], 
                   ["1000", N1000.count(0), N1000.count(1)],
                   ["5000", N5000.count(0), N5000.count(1)],
                   ["10 000", N10000.count(0), N10000.count(1)]],
                columns=['N values', '0', '1'])

# plot grouped bar chart
df.plot(x='N values', kind='bar', logy=True, title="Représentation des séquences de nombres pour différentes valeurs de N \n Loi Binomiale")
```

Loi continue : Loi normale

```python
import math

def DensiteNormale(x,mu,sigma):
    return 1/(sigma * math.sqrt(2*math.pi))*math.exp(-0.5*((x-mu)/sigma)**2)

lx=np.linspace(0,50,200)
ly=[DensiteNormale(x,15,3.2) for x in lx]
plt.plot(lx,ly, label = ' fonction de densité de la loi normale')
plt.legend(loc='upper right')
plt.show()
```

2. Ecrire un programme représentant la séquence par un histogramme pour chaque N. Puis augmenter N. Que constatez vous?

```python
N = [100,1000,5000,10000,100000]

def generate_binomial(N, n, p):
    result = []
    for i in range(N):
        count = 0
        for j in range(n):
            if random.random() < p:
                count += 1
        result.append(count)
    return result

for petitn in N :
    binomiale = generate_binomial(petitn, 50, 0.6)
    plt.hist(binomiale,bins=50,range=(1,50),density=True)
    plt.title("Histogramme de la loi binomiale pour n = " + str(petitn))
    plt.show()
```

```python
#array de 100 000 valeurs issues d'une loi normale de moyenne 5 et écart-type 2.
N = [100,1000,5000,10000,100000]

def generate_normal(n, mu, sigma):
    u = np.random.uniform(size=n)
    v = np.random.uniform(size=n)
    z = np.sqrt(-2 * np.log(u)) * np.cos(2 * np.pi * v)
    return mu + sigma * z

for n in N : 
    normale=generate_normal(n, 5, 2)
    plt.hist(normale,bins=50,density=True)
    plt.title("Histogramme de la loi normale pour n = " + str(n))
    plt.show()
```

On remarque que plus N augmente plus le diagramme se lisse. On se raproche petit à petit de la courbe de densité de loi normal.
